```python
%matplotlib inline
import pandas as pd
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
texture = pd.read_csv('texture.csv')
texture = texture.drop(['Unnamed: 0'],axis=1)
texture['texture'] = texture.texture.str.replace(',','.')
texture['texture'] = pd.to_numeric(texture['texture'])
```

```python
img = mpimg.imread('./photos/texture/1261.png')
```

```python
# fig = plt.figure(figsize=(1,10))
plt.imshow(img,cmap='gray')
```

```python
texture['category'] = pd.cut(texture.texture,bins=[-1000,0,165,275,350,460,20000],labels=['Error', 'Polido',
                                             'Macio', 'Baixa Rugosidade', 'Rugosidade Moderada',
                                                            'Alta Rugosidade'])
```

```python
f, ax = plt.subplots(figsize=(13, 4))
sns.countplot(x='category',data=texture)
```

```python
sns.distplot(texture['texture'])
```

```python
sns.distplot(texture.texture[texture.category=='Alta Rugosidade'],kde=False)
```
