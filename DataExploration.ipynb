# Data Exploration

```python
%matplotlib inline
import pandas as pd
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Import the texture data
texture = pd.read_csv('texturedataprepared.csv')
textureimages = np.load('textureimagequality20.npy')

# Assign categories names to the texture data
texture['category_label'] = pd.cut(texture.texture,bins=[-0.01,165,275,350,460,20000],labels=['Polished',
                                             'Smooth', 'Low Roughness', 'Moderate Roughness',
                                                            'High Roughness'])
```

```python
# Visualizing the data

fig = plt.figure(figsize=(40,15))
for i in range(12):
    ax = fig.add_subplot(3, 4, i + 1, xticks=[], yticks=[])
    ax.imshow((textureimages[i]))
    ax.set_title("{} | {}".format(texture['category_label'][i],texture['texture'][i]),fontsize=30)
```

```python
# Visualizing the distribution of each category in the dataset
sns.set(style="darkgrid")

f, ax = plt.subplots(figsize=(13, 4))
plt.tick_params(axis='both', which='major', labelsize=12)
sns.countplot(x='category_label',data=texture,)
plt.ylabel('Number of images',size=14)
plt.xlabel('')
plt.savefig('Texture Category Distribution.png')
```

```python
# Visualizing the distribution of the texture in the dataset

f, ax = plt.subplots(figsize=(13, 4))
sns.distplot(texture['texture'],kde=False)
plt.xlabel("Texture Index",size=14)
plt.ylabel("Number of images",size=14)
plt.savefig("Texture Index Histogram.png")
```

```python
# Visualizing the data
textureimages20 = np.load('textureimagequality20.npy')
textureimages30 = np.load('textureimagequality30.npy')
textureimages10 = np.load('textureimagequality10.npy')


fig = plt.figure(figsize=(40,15))
ax = fig.add_subplot(1, 3, 1, xticks=[], yticks=[])
ax.imshow((textureimages10[10]))
ax.set_title("Image reduction = {}%".format(10),fontsize=50)
ax = fig.add_subplot(1, 3, 2, xticks=[], yticks=[])
ax.imshow((textureimages20[10]))
ax.set_title("Image reduction = {}%".format(20),fontsize=50)
ax = fig.add_subplot(1, 3, 3, xticks=[], yticks=[])
ax.imshow((textureimages30[10]))
ax.set_title("Image reduction = {}%".format(30),fontsize=50)

plt.savefig('Image Quality.png')
```
